```python
# Rename the timestamp column in df_pupil_positions to avoid ambiguity
df_pupil_positions = df_pupil_positions.rename(columns={'timestamp': 'pupil_timestamp'})

# Reset index of both DataFrames to ensure they are comparable
df = df.reset_index(drop=True)
df_pupil_positions = df_pupil_positions.reset_index(drop=True)

# Perform the filtering operation using a loop and index comparison
filtered_pupil_positions_list = []
for i in df.index:
    mask = (df_pupil_positions['pupil_timestamp'] > df.loc[i, 'timestamp'] - 15) & \
           (df_pupil_positions['pupil_timestamp'] <= df.loc[i, 'timestamp'])

    plot_pupil_diameter(df_pupil_positions[mask], 0)
    plot_pupil_diameter(df_pupil_positions[mask], 1)


    plot_pupil_diameter_3d(df_pupil_positions[mask], 0)
    plot_pupil_diameter_3d(df_pupil_positions[mask], 1)

    filtered_pupil_positions_list.append(df_pupil_positions[mask])

# Concatenate the filtered results
filtered_pupil_positions = pd.concat(filtered_pupil_positions_list)

# Display the filtered data (optional)
filtered_pupil_positions
```

```python
def plot_pupil_diameter(filtered_pupil_positions, eye_id):
  # Filter for eye_id == 0
  filtered_pupil_positions_eye0 = filtered_pupil_positions[filtered_pupil_positions['eye_id'] == eye_id]

  # Create the plot
  plt.figure(figsize=(10, 6))
  plt.plot(filtered_pupil_positions_eye0['pupil_timestamp'], filtered_pupil_positions_eye0['diameter'])
  plt.xlabel('Timestamp')
  plt.ylabel('Pupil Diameter')
  plt.title(f'Pupil Diameter vs. Timestamp (eye_id = {eye_id})')
  plt.grid(True)
  plt.show()
```

```python
def plot_pupil_diameter_3d(filtered_pupil_positions, eye_id):
  # Filter for eye_id == 0
  filtered_pupil_positions_eye0 = filtered_pupil_positions[filtered_pupil_positions['eye_id'] == eye_id]
  filtered_pupil_positions_eye0 = filtered_pupil_positions[filtered_pupil_positions['method'] == 'pye3d 0.3.2 real-time']

  # Create the plot
  plt.figure(figsize=(10, 6))
  plt.plot(filtered_pupil_positions_eye0['pupil_timestamp'], filtered_pupil_positions_eye0['diameter_3d'])
  plt.xlabel('Timestamp')
  plt.ylabel('Pupil diameter_3d')
  plt.title(f'Pupil diameter_3d vs. Timestamp (eye_id = {eye_id})')
  plt.grid(True)
  plt.show()
```
